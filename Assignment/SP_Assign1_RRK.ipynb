```python
import IPython.display as ipd
ipd.Audio(audio_path)
```

```python
import matplotlib.pyplot as plt
import librosa.display
```

```python
plt.figure(figsize=(14,5))
librosa.display.waveshow(x,sr=sr)
```

```python
#spectogram, to see the freq vs time graph, basically to see for each sec
X=librosa.stft(x) #audio to short term fourier transform
Xdb=librosa.amplitude_to_db(abs(X))#aplitude to freq at given time
```

```python
plt.figure(figsize=(14,5))
librosa.display.specshow(Xdb,sr=sr,x_axis="time",y_axis="hz")
plt.colorbar()
```
